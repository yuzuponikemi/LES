```python
import csv
x =[]
y = []
with open('WallHF_ave_time_allSurface_inv_delt.csv', newline='') as f:

    reader = csv.reader(f)
    for row in reader:
        x.append(float(row[0]))
        y.append(float(row[1]))
                 
    print(x)
    print(y)
        

    
```

```python
a=[]
ALLHF = 0
for HF in y:
    ALLHF = ALLHF + HF
    a.append(ALLHF)
print(a)
```
